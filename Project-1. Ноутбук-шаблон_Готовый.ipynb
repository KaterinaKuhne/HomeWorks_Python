В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def work_experience(work_experience):
    if pd.isna(work_experience)==True or work_experience=='Не указано':
        return np.NaN
    else:
        work_period = work_experience.split()[0:7]
        year_variants = ['год', 'года', 'лет']
        month_variants = ['месяц', 'месяца', 'месяцев']
        total_length_work = 0
        for i in range(1,7):
            if work_period[i] in year_variants:
                total_length_work += int(work_period[i-1])*12
            elif work_period[i] in month_variants:
                total_length_work += int(work_period[i-1])
        return total_length_work


hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(work_experience)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def city_of_candidate(city_string):
    """принимает строку, вычленяет из нее город"""
    city_of_candidate = city_string.split()[0]
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород', 'Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    main_cities = ['Москва', 'Санкт-Петербург']
    if city_of_candidate in main_cities:
        return city_of_candidate
    elif city_of_candidate in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'


def check_for_not(string):
    """принимает список и проверяет есть ли в нем частица не"""
    for i in range(len(string)):
        if string[i]=='не':
            return False
        else:
            return True

def ready_to_business_trip(city_string):
    last_piece = city_string.split(',')[-1].split()
    if "командировкам" in last_piece:
        return check_for_not(last_piece)
    else:
        return False  
    
def ready_to_move(arg):
    if ('не готов к переезду' in arg) or ('не готова к переезду' in arg):
        return False
    elif 'хочу' in arg:
        return True
    else:
        return True    

hh_data['Город']=hh_data['Город, переезд, командировки'].apply(city_of_candidate)
hh_data['Готовность к командировкам'] = hh_data['Город, переезд, командировки'].apply(ready_to_business_trip)
hh_data['Готовность к переезду'] = hh_data['Город, переезд, командировки'].apply(ready_to_move)    
```

```python
hh_data=hh_data.drop(['Город, переезд, командировки'], axis=1)
```

```python
exchange_rate = pd.read_csv('/Users/katerinakuhne/Documents/МАГА МФТИ/Python/ExchangeRates.csv')
exchange_rate['close'] = exchange_rate['close']/exchange_rate['proportion']
exchange_rate['date'] = pd.to_datetime(exchange_rate['date'], dayfirst=True).dt.date
exchange_rate = exchange_rate.rename(columns={'date': 'Дата', 'currency': 'Валюта'})
exchange_rate = exchange_rate.drop(['time', 'vol', 'per', 'proportion'], axis=1)
```

```python
hh_data['Дата'] = pd.to_datetime(hh_data['Обновление резюме'], dayfirst=True).dt.date

def currency(string):
    currency = string.split()[1].split('.')[0]
    if currency=='грн':
        return 'UAH'
    elif currency=='бел':
        return 'BYN'
    elif currency=='сум':
        return 'UZS'
    else:
        return currency
    
hh_data['ЗП (руб)'] = hh_data['ЗП'].apply(lambda x: int(x.split()[0]))
hh_data['Валюта'] = hh_data['ЗП'].apply(currency)

hh_data = pd.merge(
    hh_data,
    exchange_rate,
    on=['Дата', 'Валюта'],
    how='left'
)

hh_data['close'].fillna(1, inplace=True)
hh_data['ЗП (руб)'] = hh_data['ЗП (руб)']*hh_data['close']
hh_data['Обновление резюме'] = hh_data['Дата']
```

В целом, наиболее высокие зарплатные ожидания у соискателей с высшим образованием в возрастном интервале от 38 до 46 лет. Также есть "полоса" высоких ожидаемых зарплат у соискателей с неоконченным высшим образованием, она приходится на возраста 41-49 лет. Также можно сказать, что в среднем люди с высшим образованием ожидают, что где-то до 42-46 лет их зарплаты будут расти, а в более позднем возрасте снова немного упадут. Похожая картина в категориии "неоконченное высшее". В тоже время у людей со средним и средним специальным образованием зарплатные ожидания практически не зависят от возраста.
